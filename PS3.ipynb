Step 1. Import the necessary libraries

Step 2. Import the dataset from this address.

Step 3. Assign it to a variable called users

Step 4. Discover what is the mean age per occupation

Step 5. Discover the Male ratio per occupation and sort it from the most to the least

Step 6. For each occupation, calculate the minimum and maximum ages

Step 7. For each combination of occupation and sex, calculate the mean age

Step 8. For each occupation present the percentage of women and men

```python
#1 Importing necessary libraries
import pandas as pd 
```

```python
#2 Importing and assigning to the variable users
users= pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
```

```python
users.head(8)
```

```python
# mean age per occupation
users.groupby('occupation').age.mean()
```

```python
## Male ratio per occupation and sort it from the most to the least

##total count of users


total_users = users.groupby(['occupation']).gender.count()
total_users
```

```python
##count of male by occupation
gender_male = users[users['gender']=='M']
occupation_male = gender_male.groupby(['occupation','gender']).gender.count()
occupation_male
```

```python
## Male ratio by occupation 

male_ratioperoccuption=((occupation_male/total_users)*100)
male_ratioperoccuption.sort_values(ascending=False)
```

```python
## Minimum and maximum age for each occupation
age_occupation = users.groupby('occupation')['age'].agg(['min','max'])
age_occupation
```

```python
## mean age of each combination of occupation and sex

mean_age_occu_gender = users.groupby(['occupation','gender']).age.mean()
mean_age_occu_gender
```

```python
##  Percentage of women and men for each occupation 
users_gender = users.groupby(['occupation','gender']).gender.count()
percentage_gender = (users_gender/total_users)*100
percentage_gender
```

## Question 6
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
marriage_divorce_in_US = pd.read_csv('us-marriages-divorces-1867-2014.csv')
```

```python
marriage_divorce_in_US.head(8)
```

```python
fig, ax = plt.subplots()
ax.plot(marriage_divorce_in_US.Year, marriage_divorce_in_US.Marriages_per_1000)
ax.plot(marriage_divorce_in_US.Year,marriage_divorce_in_US.Divorces_per_1000)
plt.ylabel('No. of Marriages and Divorces per capita')
plt.xlabel('Year from 1867 to 2014')
ax.set_title('Number of Marriages and Divorces per capita in the U.S. between 1867 and 2014')
ax.legend(loc='best')
```

## Question 7
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
import seaborn as sns
```

```python
year_m_d = marriage_divorce_in_US[(marriage_divorce_in_US.Year == 2000) | (marriage_divorce_in_US.Year == 1900) | (marriage_divorce_in_US.Year == 1950)]
year_m_d.plot(kind= 'bar', x='Year', y=['Marriages_per_1000','Divorces_per_1000'], label=['Marriages','Divorces'])
plt.xlabel('Year')
plt.ylabel('No of marriages and divorces per capita')
plt.title(" Number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000")
plt.legend()

```
